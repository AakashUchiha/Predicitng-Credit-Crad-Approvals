## Importing Modules

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('C:/Users/Asus/Desktop/python project/ML projects/Predicting Credit Card approvals/Book1.csv',sep=',',header=None)
df.head()
```

### Checking the unique values in each column

```python
column_list = df.columns.values.tolist()
for column_name in column_list:
    print(df[column_name].unique())
    
```

### Replacing null(?) values

```python
import numpy as np
df.replace('?',np.nan,inplace=True)
df.tail(20)
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].dtypes=='object':
        df=df.fillna(df[col].value_counts().index[0])
    else :
        df=df.fillna(df[col].mean())
```

```python
df.isnull().sum
df.dtypes
```

```python
object_columns=[]
other_columns=[]
for column_name in df.columns:
    if df[column_name].dtypes=='object' :
        object_columns.append(column_name)
    else:
        other_columns.append(column_name)

print(object_columns)
print(other_columns)
```

### Adding dummy columns for categorical variables

```python
df=pd.get_dummies(data=df,columns=df.index[object_columns],drop_first=True)
df.head()
```

### Transforming the numerical columns using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
Minmax=MinMaxScaler()
```

```python
df[other_columns]=Minmax.fit_transform(df[other_columns])
```

```python
df.head()
```

```python
X,y=df.iloc[:,:-1],df.iloc[:,-1]
```

### Splitting the data into test and train

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

### Fitting the model using Logistic Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)

y_pred=log.predict(X_test)
```

### Testing the accuracy of Logistic Regression

```python
from sklearn.metrics import classification_report 
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(log,X_test,y_test)
print(classification_report(y_pred,y_test))
```

```python
log.score(X_test,y_test)
```

```python
cross_val_score(log,X_train,y_train,cv=5).mean()
```

### Fitting the data using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
ran=RandomForestClassifier()
ran.fit(X_train,y_train)
y_pred2=ran.predict(X_test)
```

```python
plot_confusion_matrix(ran,X_test,y_test)
print(classification_report(y_pred2,y_test))
```

```python
ran.score(X_test,y_test)
```

```python
cross_val_score(ran,X_train,y_train,cv=5).mean()
```

### Fitting the data using SVM

```python
from sklearn.svm import SVC
svc=SVC(random_state=1)
svc.fit(X_train,y_train)

y_pred3=svc.predict(X_test)
```

### Testing the accuracy of SVM

```python
print(classification_report(y_pred3,y_test))
plot_confusion_matrix(svc,X_test,y_test)
```

```python
svc.score(X_test,y_test)
```

```python
cross_val_score(svc,X_train,y_train,cv=5).mean()
```

### Testing the accuracy using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train,y_train)
y_pred4=knn.predict(X_test)
```

```python
plot_confusion_matrix(knn,X_test,y_test)
print(classification_report(y_pred4,y_test))
```

```python
knn.score(X_test,y_test)
```

```python
cross_val_score(knn,X_train,y_train,cv=5).mean()
```

#### Here we see that Random Forest outperforms all other classification methods with an Accuracy of 0.91
